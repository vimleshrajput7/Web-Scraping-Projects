```python
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
```

```python
#1.find all the first string category
response=requests.get("https://stores.kohler.com/en/locations")
soup=BeautifulSoup(response.text, 'html.parser')
category=[cat.text for cat in soup.find_all(name='h3')] # loop through and stored in a list using list comprehension
print(category)
```

```python
#2.find all the city name
response=requests.get("https://stores.kohler.com/en/locations")
soup=BeautifulSoup(response.text, 'html.parser')
cities=[city.getText() for city in soup.find_all(name='h4', class_='ksslocationcards--store-city')]
print(cities)
```

```python
#extract all the store titles & links  
response=requests.get("https://stores.kohler.com/en/locations")
soup=BeautifulSoup(response.text, 'html.parser')
titles=soup.find_all(name='a', class_='ksslocationcards--store-link')
#print(title.getText())

#extract all the titles avialable stored in a list using list comprehension
all_title=[title.getText() for title in titles]
print("\n".join(all_title))

#extract links in 'a' tag
# links=[link.get('href') for link in titles]
# print(links)

#using sets stored links and it can get on webpage
links=set()
for link in titles:
  if link.get('href')!='#':
    link_text="https://stores.kohler.com/en/locations"+link.get('href')
    links.add(link_text)
print("\n".join(links))
```

```python
#2.find all the address
response=requests.get("https://stores.kohler.com/en/locations")
soup=BeautifulSoup(response.text, 'html.parser')
address=[add.text for add in soup.find_all(name='p')]
print(address)
```

```python
#2.find all the phones
response=requests.get("https://stores.kohler.com/en/locations")
soup=BeautifulSoup(response.text, 'html.parser')
contacts=[cont.getText() for cont in soup.find_all(name='a',class_='ksslocationcards--phone-link')]
print(contacts)
```

We need to combine all the list of data into dataframe and convert into csv file

```python
# combine all the list data in a dataframe
df=pd.DataFrame(list(zip(category,cities,all_title,links,address, contacts)),
             columns=['category','cities','all_title','links', 'address', 'contacts'])
```

```python
#df dataframe convert into csv
df.to_csv('kohler_data.csv')
```
